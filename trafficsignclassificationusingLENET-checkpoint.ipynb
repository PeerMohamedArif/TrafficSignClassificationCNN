# Data normalisation / Data preparation

```python
from sklearn.utils import shuffle
input_train, output_train = shuffle(input_train, output_train) # why shuffle? because the dataset is ordered in such a way a class is grouped together 
```

```python
input_train_gray = np.sum(input_train/3, axis=3, keepdims=True)
input_test_gray  = np.sum(input_test/3, axis=3, keepdims=True) # /3 IS FOR averaging the pixels since we have 3 colours  RGB  
input_validation_gray  = np.sum(input_validation/3, axis=3, keepdims=True) 
```

```python
input_train_gray.shape # will change the value from 3 to 1
```

```python
input_train_gray_normal = (input_train_gray - 128)/128 
input_test_gray_normal = (input_test_gray - 128)/128
input_validation_gray_normal = (input_validation_gray - 128)/128 
#normalisation
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, AveragePooling2D, Dense, Flatten, Dropout
from keras.optimizers import Adam
from keras.callbacks import TensorBoard

from sklearn.model_selection import train_test_split
```

```python
image_shape = input_train_gray[i].shape
```

```python
cnn_model = Sequential()

cnn_model.add(Conv2D(filters=6, kernel_size=(5, 5), activation='relu', input_shape=(32,32,1)))
cnn_model.add(AveragePooling2D(pool_size=(2, 2)))

cnn_model.add(Conv2D(filters=16, kernel_size=(5, 5), activation='relu'))
cnn_model.add(AveragePooling2D(pool_size=(2, 2)))

cnn_model.add(Flatten())

cnn_model.add(Dense(units=120, activation='relu'))

cnn_model.add(Dense(units=84, activation='relu'))

cnn_model.add(Dense(units=43, activation = 'softmax'))
```

```python
cnn_model.compile(loss ='sparse_categorical_crossentropy', optimizer=Adam(learning_rate=0.001),metrics =['accuracy']) 
# if you have two classes you can use binary but since we have 43 classes we can use sparse categorical
```

```python
history = cnn_model.fit(input_train_gray_normal,
                        output_train,
                        batch_size=500,
                        epochs=50,
                        verbose=1,
                        validation_data = (input_validation_gray_normal,output_validation))
```

```python
score = cnn_model.evaluate(input_test_gray_normal, output_test,verbose=0)
print('Test Accuracy : {:.4f}'.format(score[1]))
```

```python
history.history.keys()
```

```python
accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(accuracy))
#print(epochs)

mplt.plot(epochs, accuracy, 'bo', label='Training Accuracy')
mplt.plot(epochs, val_accuracy, 'b', label='Validation Accuracy')
mplt.title('Training and Validation accuracy')
mplt.legend()
```

```python
mplt.plot(epochs, loss, 'ro', label='Training Loss')
mplt.plot(epochs, val_loss, 'r', label='Validation Loss')
mplt.title('Training and validation loss')
mplt.legend()
mplt.show()
```

```python
# predictions for the test data
predictions = cnn_model.predict(input_test_gray_normal)
predicted_classes = np.argmax(predictions, axis=1)
#get the indices to be plotted
y_true = output_test
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_true, predicted_classes)
mplt.figure(figsize = (25,25))
sns.heatmap(cm, annot=True)
```
